```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
url = 'https://www.cga.ct.gov/2006/rpt/2006-r-0749.htm'
```

```python
driver = webdriver.Chrome()
driver.get(url)

# Locate the table using XPath
table_xpath = "/html/body/div[2]/div/table"

# Find all rows in the table
rows = driver.find_elements(By.XPATH, f"{table_xpath}/tbody/tr")

# Initialize a list to store the data
data = []

# Loop through each row and extract the data
for row in rows:
    # Extract individual cell values
    cells = row.find_elements(By.TAG_NAME, "td")
    row_data = [cell.text for cell in cells]
    
    # Append the row data to the list
    data.append(row_data)

# Print the scraped data
for row_data in data:
    print(row_data)

# Close the browser window
driver.quit()
```

```python
len(data)
```

```python
columns = ['Status Number', 'Type', 'Class', 'Name', 'Min JT', 'years', 'Max JT', 'Years', 'Min Fine', 'Max Fine']
df = pd.DataFrame(data, columns=columns)
```
