```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Finalfile.csv")
```

```python
df.head()
```

```python
# Assuming column names are "crime_news" and "label", adjust these if needed
crime_news_column = "Column_Name"
label_column = "Label"
```

```python
crime_news = df[crime_news_column].tolist()
labels = df[label_column].tolist()
```

```python
train_texts, val_texts, train_labels, val_labels = train_test_split(crime_news, labels, test_size=0.2, random_state=42)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Tokenize the input texts and convert them to input IDs and attention masks
train_encodings = tokenizer(train_texts, truncation=True, padding=True, return_tensors='pt')
val_encodings = tokenizer(val_texts, truncation=True, padding=True, return_tensors='pt')
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize label encoder
label_encoder = LabelEncoder()

# Fit label encoder on all labels to create the mapping
all_labels = labels  # Assuming 'labels' contains all the labels in your dataset
label_encoder.fit(all_labels)

# Convert labels to numerical format using label encoder
train_labels_encoded = label_encoder.transform(train_labels)
val_labels_encoded = label_encoder.transform(val_labels)

# Convert encoded labels to tensors
train_labels_tensor = torch.tensor(train_labels_encoded)
val_labels_tensor = torch.tensor(val_labels_encoded)
```

```python
# Load pre-trained BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(set(labels)))
```

```python
# Print model architecture
print(model)
```

```python
import torch
from transformers import AdamW, get_linear_schedule_with_warmup

# Define device (CPU or GPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Move model and data to device
model.to(device)
train_encodings = train_encodings.to(device)
val_encodings = val_encodings.to(device)
train_labels_tensor = train_labels_tensor.to(device)
val_labels_tensor = val_labels_tensor.to(device)

# Define optimizer and learning rate scheduler
optimizer = AdamW(model.parameters(), lr=5e-5)  # Using AdamW optimizer
num_epochs = 3  # Number of training epochs
total_steps = len(train_labels_tensor) * num_epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

# Define loss function
loss_fn = torch.nn.CrossEntropyLoss()

# Define batch size
batch_size = 8  # You can adjust this value based on your system's memory and performance
total_iterations = 1845  # Total number of iterations expected
current_iteration = 0 

# Training loop
while current_iteration < total_iterations:
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        for i in range(0, len(train_labels_tensor), batch_size):
            current_iteration += 1
            optimizer.zero_grad()
            batch_inputs = train_encodings['input_ids'][i:i+batch_size]
            batch_attention_masks = train_encodings['attention_mask'][i:i+batch_size]
            batch_labels = train_labels_tensor[i:i+batch_size].to(torch.long)
            percentage = (current_iteration / total_iterations) * 100
            print(f"\rProgress: {percentage:.2f}%", end='')  # Use carriage return and end='' to overwrite
            outputs = model(input_ids=batch_inputs, attention_mask=batch_attention_masks, labels=batch_labels)
            loss = outputs.loss
            train_loss += loss.item()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()

        avg_train_loss = train_loss / len(train_labels_tensor)

        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for j in range(0, len(val_labels_tensor), batch_size):
                batch_inputs_val = val_encodings['input_ids'][j:j+batch_size]
                batch_attention_masks_val = val_encodings['attention_mask'][j:j+batch_size]
                batch_labels_val = val_labels_tensor[j:j+batch_size].to(torch.long)
                outputs_val = model(input_ids=batch_inputs_val, attention_mask=batch_attention_masks_val, labels=batch_labels_val)
                loss_val = outputs_val.loss
                val_loss += loss_val.item()

        avg_val_loss = val_loss / len(val_labels_tensor)

        print(f'\rEpoch {epoch + 1}/{num_epochs}:')
        print(f'Training Loss: {avg_train_loss:.4f}, Validation Loss: {avg_val_loss:.4f}')
```

```python
import torch

num_correct = 0
num_total = 0
with torch.no_grad():
    for j in range(0, len(val_labels_tensor), batch_size):
        batch_inputs_val = val_encodings['input_ids'][j:j+batch_size]
        batch_attention_masks_val = val_encodings['attention_mask'][j:j+batch_size]
        batch_labels_val = val_labels_tensor[j:j+batch_size].long()  # Ensure LongTensor type
        outputs_val = model(input_ids=batch_inputs_val, attention_mask=batch_attention_masks_val, labels=batch_labels_val)
        logits = outputs_val.logits
        _, predicted_labels = torch.max(logits, 1)
        
        # Ensure both tensors have the same data type for comparison
        predicted_labels = predicted_labels.long()  # Convert predicted_labels to Long data type
        
        num_correct += (predicted_labels == batch_labels_val).sum().item()
        num_total += batch_labels_val.size(0)

# Calculate accuracy
accuracy = num_correct / num_total
print(f'Validation Accuracy: {accuracy:.2%}')
```

```python
import torch
from sklearn.metrics import classification_report

num_correct = 0
num_total = 0
all_predicted_labels = []
all_true_labels = []

with torch.no_grad():
    for j in range(0, len(val_labels_tensor), batch_size):
        batch_inputs_val = val_encodings['input_ids'][j:j+batch_size]
        batch_attention_masks_val = val_encodings['attention_mask'][j:j+batch_size]
        batch_labels_val = val_labels_tensor[j:j+batch_size].long()  # Ensure LongTensor type
        outputs_val = model(input_ids=batch_inputs_val, attention_mask=batch_attention_masks_val, labels=batch_labels_val)
        logits = outputs_val.logits
        _, predicted_labels = torch.max(logits, 1)
        
        # Ensure both tensors have the same data type for comparison
        predicted_labels = predicted_labels.long()  # Convert predicted_labels to Long data type
        
        num_correct += (predicted_labels == batch_labels_val).sum().item()
        num_total += batch_labels_val.size(0)
        
        # Collect predicted and true labels
        all_predicted_labels.extend(predicted_labels.cpu().numpy())
        all_true_labels.extend(batch_labels_val.cpu().numpy())

# Calculate accuracy
accuracy = num_correct / num_total
print("--------------------------------------")
print(f'Validation Accuracy: {accuracy:.2%}')
print("--------------------------------------")


label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

# Print the label mapping
print("Label Mapping:")
for label, encoded_value in label_mapping.items():
    print(f"{label}: {encoded_value}")
    
# Calculate other classification metrics and generate report
classification_report_str = classification_report(all_true_labels, all_predicted_labels)
print("\n--------------------------------------")
print("Classification Report:")
print(classification_report_str)
print("--------------------------------------")
```
