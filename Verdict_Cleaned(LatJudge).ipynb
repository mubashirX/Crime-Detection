```python
import pandas as pd

# Assuming df is your DataFrame with the "Name" column
# You want to extract characters inside () and 4 characters before "DEG"

# Define a function to extract characters inside parentheses
def extract_inside_parentheses(text):
    import re
    match = re.search(r'\((.*?)\)', text)
    if match:
        return match.group(1)
    else:
        return None

# Define a function to extract 4 characters before "DEG"
def extract_before_deg(text):
    import re
    match = re.search(r'(.{4})DEG', text)
    if match:
        return match.group(1)
    else:
        return None

# Apply the functions to create new columns
df['Offense'] = df['Name'].apply(extract_inside_parentheses)
df['DEG'] = df['Name'].apply(extract_before_deg)

# Print the DataFrame to see the results
print(df)
```

```python
def extract_numerical_value(text):
    if text is not None:
        # Remove non-numeric characters from the string and convert to integer
        return int(''.join(filter(str.isdigit, text)))
    else:
        return None  # Return None if the input text is None


# Apply the function to create the new column
df['DEG_NUM'] = df['DEG'].apply(extract_numerical_value)

# Print the DataFrame to see the results
print(df)
```

```python
df.to_csv("LatJudge.csv",index=True)
```
